```python
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
#Reading Hilton Stock File
hilton_csv = Path("/Users/sabrinareed/Desktop/JupyterWorkspace/covid19casedata/hilton.csv")
```

```python
#Display Hilton stock data
#Setting Date to Index / Need to find out how to drop timestamp
hilton_data = pd.read_csv(hilton_csv, header = 1, index_col="Date")
#Rename columns
hilton_data = hilton_data.rename(columns={"Close": "Hilton",})
hilton_data.head()
```

```python
#Plotting Hilton
hilton_data.hvplot()
```

```python
#Reading Carnival Stock File
carnival_csv = Path("/Users/sabrinareed/Desktop/JupyterWorkspace/covid19casedata/carnival.csv")
#Display Carnival stock data
#Setting Date to Index / Need to find out how to drop timestamp
carnival_data = pd.read_csv(carnival_csv, header = 1, index_col="Date")
#Rename columns
carnival_data = carnival_data.rename(columns={"Close": "Carnival",})
carnival_data.head()
```

```python
carnival_data.hvplot()
```

```python
#Concat
hosp_data = pd.concat([hilton_data , carnival_data], axis='columns', join='inner')
hosp_data.head()
```

```python
hosp_data.hvplot()
```
